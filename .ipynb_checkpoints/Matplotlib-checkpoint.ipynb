# IMPORTING FILES

```python
from matplotlib import pyplot as plt
```

```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
#X axis values:

x=[3,5,6,9]

#y axis values:

y=[5,8,9,10]

#Plotting the Line

plt.plot(x,y)
```

# Bar Plot

```python
#X axis values:

x=[3,5,6,9]

#y axis values:

y=[5,8,9,10]

#Plotting the box plot

plt.bar(x,y)
```

# Scatter Plot

```python
#X axis values:

x=[3,5,6,9]

#y axis values:

y=[5,8,9,10]

#Plotting Scatter Graph

plt.scatter(x,y)
```

# Labelling

```python
plt.xlabel(" X - Axis ")
plt.ylabel(" Y - Axis ")
```

# Figure Size

```python
plt.figure(figsize=(15,5))
plt.scatter(x,y)
```

# Displaying the Figure

```python
plt.show()
```
